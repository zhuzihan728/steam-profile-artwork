### Define the number of background and front body images

```python
bg_nums = 120
fb_nums = 60
```

### Preadjust size and position for front image, check new.png for results

```python
# Front Image, resize it to appropriate size
base_width = 1250

# Front Image, adjust position -left, +right
w_offset = 10

# Front Image, adjust position -top, +bottom
h_offset = 0
```

```python

# import PIL module 
from PIL import Image 
'''/* 
index logic:
bg_ind = 1, 2, 3, ..., bg_nums
fb_ind = 0, 1, 2, ..., fb_nums-1
*/'''
for bg_ind in range(1, bg_nums+1):
    if bg_ind >= fb_nums:
        fb_ind = bg_ind % fb_nums
    else:
        fb_ind = bg_ind
        
    # get index 
    fb_ind = str(fb_ind)
    bg_ind = str(bg_ind)
    
    while len(fb_ind) < 3:
        fb_ind = '0' + fb_ind
    
    while len(bg_ind) < 3:
        bg_ind = '0' + bg_ind
        
    # Front Image
    fb_name = fr"..\out\000{fb_ind}.png"
    
    # Back Image 
    bg_name = fr"bg\ezgif-frame-{bg_ind}.jpg"


    frontImage = Image.open(fb_name) 
    wpercent = (base_width / float(frontImage.size[0]))
    hsize = int((float(frontImage.size[1]) * float(wpercent)))
    frontImage = frontImage.resize((base_width, hsize))
    
    # Open Background Image 
    background = Image.open(bg_name) 

    
    # Convert image to RGBA 
    frontImage = frontImage.convert("RGBA") 
    
    # Convert image to RGBA 
    background = background.convert("RGBA") 
    
    # Adjust the front image position with respect to the background image
    
    # Calculate width
    width = (background.width - frontImage.width)//2 + w_offset
    
    # Calculate height 
    height = (background.height - frontImage.height) + h_offset
    
    # Paste the frontImage at (width, height) 
    background.paste(frontImage, (width, height), frontImage) 

    # Resize the image to 630x820
    background = background.resize((630, 820))

    print("background.size", background.size)
    # Save this image 
    background.save(f"blended/new.png",quality=1, optimize=True)

    new_image = Image.open(f"blended/new.png")
    img_left =  new_image.crop((0, 0, 506, 820))
    img_right = new_image.crop((530, 0, 630, 820))
    print("img_left.size", img_left.size)
    print("img_right.size", img_right.size)
    img_left.save(f"blended/img_left{bg_ind}.png")
    img_right.save(f"blended/img_right{bg_ind}.png")
```
